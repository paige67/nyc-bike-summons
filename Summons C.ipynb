```python
## import library
import pandas as pd
```

```python
## import data and store in dataframe + give general info
ALLdf = pd.read_csv("data-raw/BIKE-SUMMONS-C-2016-2022.csv")
ALLdf.info()
```

```python
## call last 10 rows of all years of data 

ALLdf.tail(10)
```

```python
## import more recent data and store in dataframe
recentdf = pd.read_csv("data-raw/BIKE-SUMMONS-C-2023.csv")
recentdf.info()
```

```python
## call last 10 rows of 2023 data 

recentdf.tail(10)
```

```python
## call first 10 rows of 2023 data 

recentdf.head(10)
```

```python
## what are the most common infractions in 2023
top_offenses = recentdf['OFFENSE_DESCRIPTION'].value_counts().head(10)
print(top_offenses)
```

```python
## what about the top offenses of all time (since 2016)
alltime_top_offenses = ALLdf['OFFENSE_DESCRIPTION'].value_counts().head(10)
print(alltime_top_offenses)
```

```python
## give me the the total number of infractions by race for 2023

recentdf.groupby("RACE")["OFFENSE_DESCRIPTION"].count().to_frame("Total # Offenses By Race")
```

```python
## give me the the total number of infractions by race for all time
## looking at the data on open data, they only started keeping track of race around 2016

ALLdf.groupby("RACE")["OFFENSE_DESCRIPTION"].count().to_frame("Total # Offenses By Race All Time")
```

```python
## create a subset of data for only biking on sidewalk in 2023 
## place it in a dataframe called recentdf_sidewalk
recentdf_sidewalk = recentdf.query("OFFENSE_DESCRIPTION == 'BICYCLE ON SIDEWALK'")
recentdf_sidewalk
```

```python
## now do the count by race for sidewalk infractions

sidewalk_by_race = recentdf_sidewalk.groupby("RACE").size().to_frame("count")
print(sidewalk_by_race)
```

```python
## now lets do it for all time 

alltimedf_sidewalk = ALLdf.query("OFFENSE_DESCRIPTION == 'BICYCLE ON SIDEWALK'")
alltimedf_sidewalk
```

```python
## now do the count by race for sidewalk infractions of all time

alltime_sidewalk_by_race = alltimedf_sidewalk.groupby("RACE").size().to_frame("count")
print(alltime_sidewalk_by_race)
```

```python
## do it for BICYCLE INFRACTION (COMMERCIAL) in 2023 

recentdf_commercial = recentdf.query("OFFENSE_DESCRIPTION == 'BICYCLE INFRACTION (COMMERCIAL)'")
recentdf_commercial

commercial_by_race = recentdf_commercial.groupby("RACE").size().to_frame("count")
print(commercial_by_race)
```

```python
## do it for BICYCLE INFRACTION (COMMERCIAL) all time 

alltimedf_commercial = ALLdf.query("OFFENSE_DESCRIPTION == 'BICYCLE INFRACTION (COMMERCIAL)'")
alltimedf_commercial

alltime_commercial_by_race = alltimedf_commercial.groupby("RACE").size().to_frame("count")
print(alltime_commercial_by_race)
```

```python
## do it for RECKLESS BIKE OPERATION or RECKLESS OPERATION BICYCLE all time  

alltimedf_park = ALLdf.query("OFFENSE_DESCRIPTION =='BIKE IN PARK'")
alltimedf_park

alltime_park_by_race = alltimedf_park.groupby("RACE").size().to_frame("count")
print(alltime_park_by_race)
```

```python
## query for women and race as well as what infraction women got 

alltimedf_women = ALLdf.query("SEX == 'F'")
alltimedf_women

alltime_women_by_race = alltimedf_women.groupby("RACE").size().to_frame("count")
print(alltime_women_by_race)

alltime_women_by_offense = alltimedf_women.groupby("OFFENSE_DESCRIPTION").size().to_frame("count")
print(alltime_women_by_offense)
```

```python
## what precient gives out the most tickets since 2016

top_precincts = ALLdf['PRECINCT_OF_OCCUR'].value_counts().head(10)
print(top_precincts)
```

```python
## what precient gives out the most tickets in 2023

top_precincts_recent = recentdf['PRECINCT_OF_OCCUR'].value_counts().head(10)
print(top_precincts_recent)
```

```python
## top boros for all time and 2023? 

top_boro = ALLdf['BORO'].value_counts().head(5)
print(top_boro)

top_boro_recent = recentdf['BORO'].value_counts().head(5)
print(top_boro_recent)
```

```python
## top age groups for all time and 2023? 

top_age = ALLdf['AGE_GROUP'].value_counts().head(5)
print(top_age)

top_age_recent = recentdf['AGE_GROUP'].value_counts().head(5)
print(top_age_recent)
```
