```python
## import library
import pandas as pd
```

```python
## import data and store in dataframe + give general info
ALLdf = pd.read_csv("data-raw/BIKE-SUMMONS-C-2016-2024.csv")
ALLdf.info()
```

```python
## call the earliest 5 rows of all (2016 - 2024) years of data 
ALLdf.tail(5)
```

```python
## import more recent data and store in dataframe
recentdf = pd.read_csv("data-raw/BIKE-SUMMONS-C-2024.csv")
recentdf.info()
```

```python
## call the earliest 5 rows of 2024 data 

recentdf.tail(5)
```

```python
## call the latest 5 rows of 2024 data 

recentdf.head(5)
```

```python
## import more 2023 data and store in dataframe --> in case we want to compare 2023 to 2024 
lastyeardf = pd.read_csv("data-raw/BIKE-SUMMONS-C-2023.csv")
lastyeardf.info()
```

```python
## call the last rows of 2023 data 

lastyeardf.head(5)
```

```python
## what are the most common infractions in 2024
top_offenses = recentdf['OFFENSE_DESCRIPTION'].value_counts().head(5)
print(top_offenses)
```

```python
## what about the top offenses of all time (since 2016)
alltime_top_offenses = ALLdf['OFFENSE_DESCRIPTION'].value_counts().head(5)
print(alltime_top_offenses)
```

```python
## give me the the total number of infractions by race for 2024

recentdf.groupby("RACE")["OFFENSE_DESCRIPTION"].count().to_frame("Total # Offenses By Race")
```

```python
## now lets do the same for 2023

lastyeardf.groupby("RACE")["OFFENSE_DESCRIPTION"].count().to_frame("Total # Offenses By Race")
```

```python
## give me the the total number of infractions by race for all time
## looking at the data on open data, they only started keeping track of race around 2016

ALLdf.groupby("RACE")["OFFENSE_DESCRIPTION"].count().to_frame("Total # Offenses By Race All Time")
```

```python
## create a subset of data for only biking on sidewalk in 2024
## place it in a dataframe called recentdf_sidewalk
recentdf_sidewalk = recentdf.query("OFFENSE_DESCRIPTION == 'BICYCLE ON SIDEWALK'")
recentdf_sidewalk
```

```python
## now do the count by race for sidewalk infractions

sidewalk_by_race = recentdf_sidewalk.groupby("RACE").size().to_frame("count")
print(sidewalk_by_race)
```

```python
## now lets do it for all time 

alltimedf_sidewalk = ALLdf.query("OFFENSE_DESCRIPTION == 'BICYCLE ON SIDEWALK'")
alltimedf_sidewalk
```

```python
## now do the count by race for sidewalk infractions of all time

alltime_sidewalk_by_race = alltimedf_sidewalk.groupby("RACE").size().to_frame("count")
print(alltime_sidewalk_by_race)
```

```python
## do it for BICYCLE INFRACTION (COMMERCIAL) in 2024

recentdf_commercial = recentdf.query("OFFENSE_DESCRIPTION == 'BICYCLE INFRACTION (COMMERCIAL)'")
recentdf_commercial

commercial_by_race = recentdf_commercial.groupby("RACE").size().to_frame("count")
print(commercial_by_race)
```

```python
## do it for BICYCLE INFRACTION (COMMERCIAL) all time 

alltimedf_commercial = ALLdf.query("OFFENSE_DESCRIPTION == 'BICYCLE INFRACTION (COMMERCIAL)'")
alltimedf_commercial

alltime_commercial_by_race = alltimedf_commercial.groupby("RACE").size().to_frame("count")
print(alltime_commercial_by_race)
```

```python
## do it for BIKE IN PARK all time  

alltimedf_park = ALLdf.query("OFFENSE_DESCRIPTION =='BIKE IN PARK'")
alltimedf_park

alltime_park_by_race = alltimedf_park.groupby("RACE").size().to_frame("count")
print(alltime_park_by_race)
```

```python
## query for women and race as well as what infraction women got 

alltimedf_women = ALLdf.query("SEX == 'F'")
alltimedf_women

alltime_women_by_race = alltimedf_women.groupby("RACE").size().to_frame("count")
print(alltime_women_by_race)

alltime_women_by_offense = alltimedf_women.groupby("OFFENSE_DESCRIPTION").size().to_frame("count")
print(alltime_women_by_offense)
```

```python
## what precient gives out the most tickets since 2016

top_precincts = ALLdf['PRECINCT_OF_OCCUR'].value_counts().head(10)
print(top_precincts)
```

```python
## what precient gives out the most tickets in 2024

top_precincts_recent = recentdf['PRECINCT_OF_OCCUR'].value_counts().head(10)
print(top_precincts_recent)
```

```python
## top boros for all time and 2024? 

top_boro = ALLdf['BORO'].value_counts().head(5)
print(top_boro)

top_boro_recent = recentdf['BORO'].value_counts().head(5)
print(top_boro_recent)
```

```python
## top age groups for all time and 2024? 

top_age = ALLdf['AGE_GROUP'].value_counts().head(5)
print(top_age)

top_age_recent = recentdf['AGE_GROUP'].value_counts().head(5)
print(top_age_recent)
```

```python
## calculate percentages for summons by race for 2024 

recent_race_totals = recentdf.groupby("RACE")["OFFENSE_DESCRIPTION"].count()
recent_race_percents = (recent_race_totals / recent_race_totals.sum() * 100).round(1)
recent_race_percents.to_frame("Percent of Total Summons by Race (2024)")
```

```python
## calculate percentages for summons by race for 2023

last_year_race_totals = lastyeardf.groupby("RACE")["OFFENSE_DESCRIPTION"].count()
last_year_race_percents = (last_year_race_totals / last_year_race_totals.sum() * 100).round(1)
last_year_race_percents.to_frame("Percent of Total Summons by Race (2023)")
```

```python
## calculate percentages for summons by race from 2016 - 2024 

all_time_race_totals = ALLdf.groupby("RACE")["OFFENSE_DESCRIPTION"].count()
all_time_race_percents = (all_time_race_totals / all_time_race_totals.sum() * 100).round(1)
all_time_race_percents.to_frame("Percent of Total Summons by Race")
```

```python
## this data is updated quarterly, here is an update with the 2025 first quarter data 

firstqdf = pd.read_csv("data-raw/BIKE-SUMMONS-C-2025.csv")
ALLdf.info()
```

```python
## here are the most recent infractions that we have in our data 

firstqdf.tail(5)
```

```python
## what are the most common infractions in 2025
firstq_top_offenses = firstqdf['OFFENSE_DESCRIPTION'].value_counts().head(5)
print(firstq_top_offenses)
```

```python
## give me the the total number of infractions by race for the first quarter of 2025

firstqdf.groupby("RACE")["OFFENSE_DESCRIPTION"].count().to_frame("Total # Offenses By Race")
```

```python
## let's look at sidewalk infractions in 2025


firstqdf_sidewalk = firstqdf.query("OFFENSE_DESCRIPTION =='BICYCLE ON SIDEWALK'")
firstqdf_sidewalk

firstqdf_sidewalk_by_race = firstqdf_sidewalk.groupby("RACE").size().to_frame("count")
print(firstqdf_sidewalk_by_race)
```
